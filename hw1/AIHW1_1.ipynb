```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```
def f1(x):
  return (x**3) - (23*x) + 1

def f3(x, y):
  return (np.sin(x**2 - y**2))
```

```
# returns x values of found local minimums
def findMin(xs, ys):
  mins = []
  first = xs[0]
  last = xs[len(xs)-1]
  for i in xs:
    if i != first and i != last:
      if ys[i-1] > ys[i] and ys[i+1] > ys[i]:
        mins.append(xs[i])
  return mins
```

```
# returns x values of found local maximums
def findMax(xs, ys):
  maxs = []
  first = xs[0]
  last = xs[len(xs)-1]
  for i in xs:
    if i != first and i != last:
      if ys[i-1] < ys[i] and ys[i+1] < ys[i]:
        maxs.append(xs[i])
  return maxs
```

```
def findMin3d(xs, ys, zs):
  mins = []
  
  for x in range(len(xs)):
    if x != 0 and x != len(xs)-1:
      for y in range(len(ys)):
        if y != 0 and y != len(ys)-1:
          a = zs[x, y]
          if a < zs[x-1, y] and a < zs[x+1, y] and a < zs[x, y-1] and a < zs[x, y+1] and a < zs[x-1, y-1] and a < zs[x+1, y-1] and a < zs[x+1, y-1] and a < zs[x-1, y+1]:
            mins.append((xs[x,y], ys[x,y], zs[x,y]))
            
  return mins
          
```

```
def findMax3d(xs, ys, zs):
  maxs = []
  
  for x in range(len(xs)):
    if x != 0 and x != len(xs)-1:
      for y in range(len(ys)):
        if y != 0 and y != len(ys)-1:
          a = zs[x, y]
          if a > zs[x-1, y] and a > zs[x+1, y] and a > zs[x, y-1] and a > zs[x, y+1] and a > zs[x-1, y-1] and a > zs[x+1, y-1] and a > zs[x+1, y-1] and a > zs[x-1, y+1]:
            maxs.append((xs[x,y], ys[x,y], zs[x,y]))
           
  return maxs
```

```
x1 = [x1 for x1 in range(-10, 11)]
y1 = [f1(x1) for x1 in x1]

minx1 = findMin(x1, y1)
miny1 = [f1(minx1) for minx1 in minx1]

maxx1 = findMax(x1, y1)
maxy1 = [f1(maxx1) for maxx1 in maxx1]

plt.title("f1(x) = x^3 - 23x + 1")
plt.plot(x1, y1, '-', minx1, miny1, 'bs', maxx1, maxy1, 'g^')
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X, Y, Z = axes3d.get_test_data(0.05)

min3d = findMin3d(X,Y,Z)
minx = [min3d[0][0]]
miny = [min3d[0][1]]
minz = [min3d[0][2]]

max3d = findMax3d(X, Y, Z)
maxx = [max3d[0][0]]
maxy = [max3d[0][1]]
maxz = [max3d[0][2]]


ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.plot(minx, miny, minz, marker = '.', markersize = 15, alpha = 1, zorder = 100)
ax.plot(maxx, maxy, maxz, marker = '.', markersize = 15, alpha = 1, zorder = 100)

plt.show()
```
